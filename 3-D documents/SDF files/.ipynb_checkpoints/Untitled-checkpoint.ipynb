```python
from thermo.chemical import Chemical
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.formula.api as sm
import statsmodels.api as sma

matplotlib.style.use('ggplot')

import csv
import cirpy
```

```python
comps = open('database.csv','r')
data=open('data1.csv','w')
readCSV = csv.reader(comps, delimiter=',')
r = csv.writer(data)
r.writerow(['Name','COMPOUND','SOLUBILITY','MW','STRUCTURE'])
for row in readCSV:
    str=""
    try:
        name = row[0]
        c = open(name+'.sdf','r')
        num_lines = sum(1 for line in open(name+'.sdf'))
        readMe = open(name+'.sdf','r').readlines()
        for l in range(num_lines-4):
             n = readMe[l+2].strip('\n')
             n = n.strip('\t')
             str+=n+' '
        d=[name,row[1],row[2],row[3],str]
        r.writerow(d)
        c.close()
    except Exception as e:
        
        pass
comps.close()
data.close()
```
